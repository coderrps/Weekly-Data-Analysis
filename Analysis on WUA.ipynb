- New Users = current - Previous
- Resurrected User = current and previous
- Churned User = previous - current
- Retained User = current and previous
- Quick ratio = (new_user + resurrected_user) / churned

```python
def calculate_new_users(current_week, previous_week):
    new_users = set(current_week) - set(previous_week)
    return len(new_users)

def calculate_resurrected_users(current_week, previous_week):
    resurrected_users = set(current_week) & set(previous_week)
    return len(resurrected_users)

def calculate_churned_users(current_week, previous_week):
    churned_users = set(previous_week) - set(current_week)
    return len(churned_users)

def calculate_retained_users(current_week, previous_week):
    retained_users = set(current_week) & set(previous_week)
    return len(retained_users)

def calculate_quick_ratio(retained_users, total_previous_week):
    if total_previous_week == 0:
        return 0.0
    else:
        return retained_users / total_previous_week
```

```python
weeks = df.columns
metrics = []
```

```python
previous_week_data = set(df[weeks[0]])
```

```python
previous_week_data
```

```python
for i in range(1, len(weeks)):
    current_week_data = set(df[weeks[i]])
    
    new_users = calculate_new_users(current_week_data, previous_week_data)
    resurrected_users = calculate_resurrected_users(current_week_data, previous_week_data)
    churned_users = calculate_churned_users(current_week_data, previous_week_data)
    retained_users = calculate_retained_users(current_week_data, previous_week_data)
    total_previous_week = len(previous_week_data)
    quick_ratio = calculate_quick_ratio(retained_users, total_previous_week)
    
 
    metrics.append({
        'Week': weeks[i],
        'New Users': new_users,
        'Resurrected Users': resurrected_users,
        'Churned Users': churned_users,
        'Retained Users': retained_users,
        'Quick Ratio': quick_ratio
    })
    
    
    previous_week_data = current_week_data
```

```python
metrics_df = pd.DataFrame(metrics)
```
